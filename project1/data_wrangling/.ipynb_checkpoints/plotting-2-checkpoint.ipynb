# Step 2: Raw Data Visualization
## Import raw data

```python
import pickle
data = pickle.load( open( "df.pkl", "rb" ) )
```

## Import necessary packages

```python
!pip install ggplot
import pandas as pd
import numpy as np
from ggplot import *
import matplotlib as plt
import datetime
```

## Convert data to dataframe

```python
dates , deaths = [] , []
for a , b in data:
    dates.append(a)
    deaths.append(b)

dates = [pd.to_datetime(d) for d in dates]

df = pd.DataFrame(list(zip(dates, deaths)), index = dates, columns = ('Dates', 'Deaths'))

dates = df.loc['2017-01-01':'2018-05-30']
deaths = df['Deaths'] > 0

df = df[deaths]
df = df.loc['2017-01-01':'2018-05-30']
```

## Scatter plot of raw data for 2017-2018 deaths in Puerto Rico

```python
g = (ggplot(aes(x = 'Dates', y = 'Deaths'), data = df) +\
geom_point(alpha = 0.5, colour = "grey") +\
ggtitle("Raw Data Dates vs Deaths 2017-2018 (Python)") +\
scale_x_date(labels = "%b", date_breaks = "1 months", date_minor_breaks = "1 years") +\
scale_y_continuous(limits = (0, 150),
        labels = labs(50, 150, 50),
        breaks = range(0, 150, 50)) +\
geom_vline(x = [pd.to_datetime(datetime.date(2017,9,20))], linetype = "dashed", color = "grey"))
print(g)
```
